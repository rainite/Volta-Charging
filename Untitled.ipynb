## Display all of the stations on a map

```python
import requests
headers = {
  'Accept': 'application/json'
}

r = requests.get('https://api.voltaapi.com/v1/stations', params={

}, headers = headers)

print (r.json()[0])
```

```python
lon = []
lat = []
for i in r.json():
    x = i['location']['coordinates'][0]
    y = i['location']['coordinates'][1]
    lon.append(x)
    lat.append(y)
print(len(lon))
print(len(lat))
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

KM = 1000.
clat = 39.3
clon = -94.7333
wid = 5500 * KM
hgt = 3500 * KM
map = Basemap(width=wid, height=hgt, rsphere=(6378137.00,6356752.3142),
            resolution='i', area_thresh=2500., projection='lcc',
            lat_1=38.5, lat_2=38.5, lat_0=clat, lon_0=clon)
 
map.drawcoastlines()
map.drawcounties()
map.drawcountries()
map.drawstates()
map.fillcontinents(color='#FAFAFA')
map.drawmapboundary()
 

lons = []
lats = []
for i in r.json():
    x = i['location']['coordinates'][0]
    y = i['location']['coordinates'][1]
    lons.append(x)
    lats.append(y)

x,y = map(lons, lats)
map.plot(x, y, 'bo', markersize=5)
plt.figure(figsize = (20,40))
plt.show()
```

## Implement a search by name function

```python
def find(input):
    for i in r.json():
        name = i['name']
        if (name == input):

            x = i['location']['coordinates'][0]
            y = i['location']['coordinates'][1]
            return([x,y])
```
